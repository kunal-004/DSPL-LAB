```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv("C:\\Users\\kunal rajesh pitale\\Downloads\\flight2\\flights.csv")
data
```

```python
# Using K-Means Clustering
x = data.iloc[:,5:7] # 1st for rows and second for columns
x
```

```python
x= x.dropna()

kmeans = KMeans(2) # Creating two clusters
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['DISTANCE'],data_with_clusters['price'],c=data_with_clusters['Clusters'],cmap='rainbow')
```
